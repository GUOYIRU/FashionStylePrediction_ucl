```python
import numpy as np
import pandas as pd
from gensim.models import FastText
from gensim.models.word2vec import Word2Vec
from gensim.models import KeyedVectors

import time
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import warnings
warnings.filterwarnings('ignore',category=pd.io.pytables.PerformanceWarning)
```

```python
datapath1=''
datapath2=''
# datapath1='../data/modified/'
# datapath2='../data/embeddings/'
```

```python
df=pd.read_hdf(datapath1+'df.h5', 'table')
weight_df=pd.read_hdf(datapath2+'tfidf_weights_df.h5', 'table')
```

```python
desc=df['comb_desc'].apply(lambda x: x.split(' '))
```

```python
df['comb_desc'][df['comb_desc'].str.contains('gothic')==True]
```

```python
start_time = time.time()
model = FastText(desc, 
                 size=50, 
                 window=5,
                 iter=20,
                 min_count=5,
                 word_ngrams=2,
                 workers=16,
                 bucket=600,
                 seed=2018,
                 sample=0.02,
                 sg=1)# 1: skip-gram, otherwise: CBOW
print("--- %.2f minutes ---" % ((time.time() - start_time)/60))
```

```python
fasttext_vectors = model.wv
model.save('model/fasttext/ft_model_50')
fasttext_vectors.save('model/fasttext/ft_embeddings_50')

model = FastText.load('model/fasttext/ft_model_50')
fasttext_vectors = KeyedVectors.load('model/fasttext/ft_embeddings_50')
```

```python
tfidf_dict = dict(zip(weight_df.term, weight_df.weight))
```

```python
def tfidf_mapping(desc, word_vectors, dims):
    desc_vec=[]
    for s in desc:
        try:
            desc_vec.append(tfidf_dict[s]*word_vectors[s])
        except:
            pass
    if len(desc_vec)==0:
        return np.zeros(shape=(dims), dtype='float32')
    else:
        return np.array(desc_vec).mean(axis=0)
```

```python
%%time
weighted_desc_embedding=desc.apply(lambda row: tfidf_mapping(row, fasttext_vectors, 200))
```

```python
df['tf_ft_desc_embedding']=weighted_desc_embedding
```
